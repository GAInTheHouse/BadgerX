```
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Sun Oct 11 01:38:08 2020

@author: shreyanssaraogi
"""
from netCDF4 import Dataset
import numpy as np
import pandas as pd

for date in dates: 

  print(date)
  df = pd.DataFrame()
  datafiles = glob.glob('/content/data3/*2018{}*'.format(date))

  if datafiles == []:
    print(date + " is not available.")


  ts_total = []
  latarr_total = []
  lonarr_total = []
  temparr_total =[]
  humidityarr_total =[]
  vwindarr_total = []
  uwindarr_total = []
  cloud_water_total = []


  for j in range(len(datafiles)):
      datafile_name = datafiles[j]
      data = Dataset(datafile_name, 'r')


      # print(data.variables.keys())
      # print(data.variables)
      ct = 0
      latarr= []
      lonarr = []
      temparr =[]
      humidityarr =[]
      vwindarr = []
      uwindarr = []
      preparr= [] 

      
      ct = 0
      i = 0 


      
      
      
      for lat_grd, lon_grd, temp_grd, hmd_grd, v_grd, u_grd, prep_grd in zip(data.variables['gridlat_221'][:], data.variables['gridlon_221'][:], data.variables['TMP_221_ISBL'][-1][:],data.variables['SPF_H_221_ISBL'][-1][:], data.variables['V_GRD_221_ISBL'][-1][:], data.variables['U_GRD_221_ISBL'][-1][:],data.variables['CLWMR_221_ISBL'][-1][:]):
          for lat, lon, tmp, hmd, vwind, uwind, prep in zip(lat_grd, lon_grd, temp_grd, hmd_grd, v_grd, u_grd, prep_grd): 
              if((lat<43.0) and (lat>33.0) and (lon<-116.0) and (lon>-126.0) ):
                  latarr.append(lat)
                  lonarr.append(lon)
                  temparr.append(tmp)
                  humidityarr.append(hmd)
                  vwindarr.append(vwind)
                  uwindarr.append(uwind)
                  preparr.append(prep)



              ct += 1
              
      timestamp = datafile_name.split(".")[-3]
      #print(timestamp)
      
      ts_total += [timestamp] * len(latarr)
      latarr_total += latarr
      lonarr_total += lonarr
      temparr_total += temparr
      humidityarr_total += humidityarr
      vwindarr_total += vwindarr
      uwindarr_total += uwindarr
      cloud_water_total += preparr


      
      #print(ct)
      
  df["timestamp"]  = ts_total
  df["latitude"] = latarr_total
  df["longitude"] = lonarr_total
  df["temperature"] = temparr_total
  df["humidity"] = humidityarr_total
  df["v_component"] = vwindarr_total
  df["u_component"] = uwindarr_total
  df["precipitation"] = cloud_water_total

  from tabulate import tabulate
  #print(tabulate(df.head(n=5)))

  df.to_csv("2018{}.csv".format(date))


```
